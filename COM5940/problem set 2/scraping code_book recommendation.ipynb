```python
%%time
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup 
from time import sleep
from random import randint 
from time import time

header = ['书名','链接','信息']
data = []

for i in range(1,4):
    quote_page = requests.get('https://book.douban.com/top250?start='+str((i-1)*25))
    sleep(randint(8,15))
    print("\n***** Page " + str(i) +" in action *****")
    soup = BeautifulSoup(quote_page.content,'html.parser')
                              
    for article in soup.find_all("tr", attrs={"class":"item"}): 
        page_no = str(i)
        
        all_a = article.find_all('a')
        url = all_a[0].get('href')
        
        title = all_a[1].text.encode('utf-8').strip()
        decoded_title = title.decode('utf-8')
        
        all_p = article.find_all('p')
        details = all_p[0].text.encode('utf-8').strip()
        decoded_details = details.decode('utf-8')

        print(decoded_title)
        print(url)
        print(decoded_details)
        data.append(( decoded_title, url, decoded_details))
                              
df = pd.DataFrame(data, 
    columns = header
)
df.to_csv('book_recommendation.csv', sep='\t', encoding='utf-8')
```
